# E01: Leer ficheros de intermagnet

```python
# importamos las librerías que vamos a necesitar
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime
```

```python
# trabajaremos con el fichero siguiente:
    # observatorio: hermanus (HER)
    # día: 02-10-2024
    # tipo: quasi-definitive (q)
    # resolución: minutos (min)
filename='data/HER/her20241002qmin.min'
```

```python
# leemos el fichero
names=['diahora','doy','c1','c2','c3','c4']
colspecs=[(0,23), (25,27), (32,40), (42,50), (52,60), (62,70)]
df=pd.read_fwf(filename,header=21,names=names, colspecs=colspecs)
```

```python
# creamos las variables que nos servirán luego
tiempo=pd.to_datetime(df.diahora.apply(str), format = '%Y-%m-%d %H:%M:%S.%f')
dat1=df.c1   # componente 1
dat2=df.c2   # componente 2
dat3=df.c3   # componente 3


# comprobar tamaño de las variables
print(dat1.size)
print(dat2.size)
print(dat3.size)
print(tiempo.size)
```
