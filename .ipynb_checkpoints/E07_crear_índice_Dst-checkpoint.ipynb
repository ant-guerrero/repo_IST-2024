## E07: Crear una versión preliminar del índice Dst

```python
# importamos las librerías que vamos a necesitar
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime
import numpy as np

# nuestra librería
import ist_lib as ist

folder='data/'
month='202410'

# observatorios usados para el índice Dst
obs=['her','hon','kak','sjg']

```

# Obtenemos todos los días del mes en la variable "dat"

```python
# días a observar: todo el mes '01' - '31'
qs=pd.Series(range(1,32))
qs=pd.Series.tolist(qs.apply(lambda x: '{0:0>2}'.format(x)))
print(qs)

# variable para guardar todos los datos (1440 minutos) de los cuatro observatorios,
# los días necesarios
dat = np.zeros([len(obs),len(qs),1440])

# leer todos los días en calma del mes para cada observatorio (sólo la componente H (c1))
for o in obs:
    for q in qs:
        filename=folder + o + '/' + o + month + q + 'qmin.min'
        df=ist.read_im(filename=filename)
        data=df['data']
        dat[obs.index(o)][qs.index(q)][:]=data.c1
```

# Obtenemos los días en calma en la variable "dat_quiets"

```python
# días en calma para el mes de octubre de 2024
qs=['25','13','21','01','02']

# variable para guardar todos los datos (1440 minutos) de los cuatro observatorios,
# los días necesarios
dat_quiets = np.zeros([len(obs),len(qs),1440])

# leer todos los días en calma del mes para cada observatorio (sólo la componente H (c1))
for o in obs:
    for q in qs:
        filename=folder + o + '/' + o + month + q + 'qmin.min'
        df=ist.read_im(filename=filename)
        data=df['data']
        dat_quiets[obs.index(o)][qs.index(q)][:]=data.c1
```

# Observemos los datos sin la eliminación de días en calma

```python
# Todos los días del mes para los cuatro observatorios
for io in range(4): # itera entre observatorios
    plt.plot(dat[io][:][:].reshape(-1));
    plt.title(obs[io].upper())
    plt.show()
```

```python
# Creamos el día en calma promedio para cada observatorio
quiets = np.zeros([len(obs),1440])
sq = np.zeros([len(obs),31,1440]) # variable que contiene para los cuatro obs, 31 días de 1440 minutos

for io in range(4): # itera entre observatorios
    quiets[io][:]=np.mean(dat_quiets[io][:][:],axis=0)
    for i in range(31): # Replicamos el día en calma promedio para todo el mes en los cuatro observatorios
        sq[io][i][:] = quiets[io][:]
    loc_ind=dat[io][:][:]-sq[io][i][:] # Creamos el índice restando a los datos los días en calma
    
    plt.plot(loc_ind.reshape(-1));
    plt.title(obs[io].upper())
    plt.show()
```

# Creamos un índice Dst a partir de los cuatro índices locales

```python
loc_inds = dat[:][:][:] - sq[:][:][:]  # Guardamos los índices locales en una variable
dst_ist = np.nanmean(loc_inds, axis=0)

fig, ax = plt.subplots(figsize=(15, 3))
plt.ylim(-500, 100)

ax.plot(dst_ist.reshape(-1), label="DST IST")

num_minutes = len(dst_ist.reshape(-1))  # Número total de datos
num_days = num_minutes // 1440

# Crear ubicaciones de ticks para cada día
x_ticks = np.arange(0, num_days * 1440 + 1, 1440)  # Ticks en los bordes de cada día
x_labels_positions = x_ticks[:-1] + 720  # Posiciones centradas entre los ticks
x_labels = [str(day) for day in range(1, num_days + 1)]  # Etiquetas para cada día

# Configurar ticks principales (pero no mostrar sus etiquetas)
ax.set_xticks(x_ticks)  # Posiciones de los ticks en los bordes de cada día

# Configurar etiquetas centradas en los ticks menores
ax.set_xticks(x_labels_positions, minor=True)  # Etiquetas centradas como 'minor ticks'
ax.set_xticklabels(x_labels, minor=True)  # Mostrar las etiquetas solo para los ticks menores

# Ocultar las etiquetas de los ticks principales
ax.set_xticklabels([])  # No mostrar etiquetas para los ticks mayores

# Ocultar los ticks menores y su grid
ax.tick_params(axis='x', which='minor', length=0)  # Ocultar las marcas menores
ax.grid(which="minor", linestyle="", linewidth=0)  # Desactivar la cuadrícula menor

# Configurar los límites del eje X para que coincidan exactamente con los datos
ax.set_xlim(0, num_minutes - 1)  # Límite exacto basado en los datos

# Configurar el rango del eje Y
y_ticks = np.arange(-500, 101, 100)  # De -500 a 100 con pasos de 100
ax.set_yticks(y_ticks)

# Añadir una cuadrícula solo para los ticks principales
ax.grid(which="major", linestyle="--", linewidth=0.7, alpha=0.7)

# Añadir título y etiquetas
ax.set_title("DST IST")
ax.set_xlabel("Días del mes")
ax.set_ylabel("Valores DST")

# Mostrar leyenda
ax.legend()

# Mostrar la gráfica
plt.show()
```

## A04: Compara el método oficial para obtener el Dst en la web de Kyoto WDC y comenta las diferencias más significativas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#################################################################
#######     DST IST
#################################################################
loc_inds = dat[:][:][:] - sq[:][:][:]  # Guardamos los índices locales en una variable
dst_ist = np.nanmean(loc_inds, axis=0)

#################################################################
### DST original
#################################################################

# Ruta del archivo con los datos de Dst oficiales
file_path = 'data/dst_original_october_2024.txt'  # Cambia esto por la ruta de tu archivo .txt

# Leer solo las 24 columnas de los valores horarios (4 caracteres por columna)
# Usamos `pd.read_fwf` con `colspecs` para leer solo las columnas de interés
colspecs = [(20 + i * 4, 24 + i * 4) for i in range(24)]  # Cada columna tiene un ancho de 4 caracteres

# Leer el archivo con las longitudes fijas de las columnas
df = pd.read_fwf(file_path, colspecs=colspecs, header=None, encoding='utf-8')

# Convertir las columnas a valores numéricos, reemplazando 9999 por NaN (datos faltantes)
df = df.apply(pd.to_numeric, errors='coerce')

# Aplanar los valores horarios en una sola lista (un valor por cada hora)
dst = df.values.flatten()

#################################################################
### Resample DST to minute resolution
#################################################################

# Crear una serie temporal para DST con resolución por minutos
minutes_dst = np.repeat(dst, 60)  # Repetir cada valor horario 60 veces para obtener los minutos

#################################################################
### PLOTS
#################################################################

# Crear la figura y el eje
fig, ax = plt.subplots(figsize=(15, 3))
plt.ylim(-500, 100)

# Dibujar la variable dst_ist (con resolución de minutos)
ax.plot(dst_ist.reshape(-1), label="DST IST")

# Dibujar la variable dst (con resolución de minutos, resampleada)
ax.plot(minutes_dst, label="DST Original")  # Línea discontinua para dst

num_minutes = len(dst_ist.reshape(-1))  # Número total de datos
num_days = num_minutes // 1440

# Crear ubicaciones de ticks para cada día
x_ticks = np.arange(0, num_days * 1440 + 1, 1440)  # Ticks en los bordes de cada día
x_labels_positions = x_ticks[:-1] + 720  # Posiciones centradas entre los ticks
x_labels = [str(day) for day in range(1, num_days + 1)]  # Etiquetas para cada día

# Configurar ticks principales (pero no mostrar sus etiquetas)
ax.set_xticks(x_ticks)  # Posiciones de los ticks en los bordes de cada día

# Configurar etiquetas centradas en los ticks menores
ax.set_xticks(x_labels_positions, minor=True)  # Etiquetas centradas como 'minor ticks'
ax.set_xticklabels(x_labels, minor=True)  # Mostrar las etiquetas solo para los ticks menores

# Ocultar las etiquetas de los ticks principales
ax.set_xticklabels([])  # No mostrar etiquetas para los ticks mayores

# Ocultar los ticks menores y su grid
ax.tick_params(axis='x', which='minor', length=0)  # Ocultar las marcas menores
ax.grid(which="minor", linestyle="", linewidth=0)  # Desactivar la cuadrícula menor

# Configurar los límites del eje X para que coincidan exactamente con los datos
ax.set_xlim(0, num_minutes - 1)  # Límite exacto basado en los datos

# Configurar el rango del eje Y
y_ticks = np.arange(-500, 101, 100)  # De -500 a 100 con pasos de 100
ax.set_yticks(y_ticks)

# Añadir una cuadrícula solo para los ticks principales
ax.grid(which="major", linestyle="--", linewidth=0.7, alpha=0.7)

# Añadir título y etiquetas
ax.set_title("DST IST vs DST Original")
ax.set_xlabel("Días del mes de octubre de 2024")
ax.set_ylabel("Valores DST")

# Mostrar leyenda
ax.legend()

# Mostrar la gráfica
plt.show()
```
